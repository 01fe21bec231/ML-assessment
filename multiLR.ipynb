```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data1=pd.read_csv('/content/sample_data/california_housing_train.csv')
print(data1)
x2=data1[['housing_median_age','total_rooms','total_bedrooms']]
y2=data1[['population']]

print("x1",x2)
print("y1",y2)
plt.plot(x2,y2,color='g',marker="*")
plt.xlabel("independent")
plt.ylabel("dependet")
plt.show()

slr=LinearRegression()
slr.fit(x2,y2)
pred=slr.predict(x2)

mse=mean_squared_error(y2,pred)
plt.plot(x2,y2,color='b',marker="*")
plt.plot(x2,pred,color='r',marker='*')
plt.title("linear-multi")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

print("mse:",mse)
print("intercept",slr.intercept_)
print("slope:",slr.coef_)
```
