FIND THE LINEAR REGRSSION USING RANDOM GENERATOR

A] Instead of directly giving values, randomly generate x and y using random module and do linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
x=np.random.randint(100,1000,15).reshape((-1,1))
y=np.random.randint(100,1000,15)
#same data in both x and y
plt.scatter(x,y)
plt.title(" linear regression")
plt.xlabel("X- AXIS")
plt.ylabel("Y- AXIS")
plt.show()

#preparing for the process
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
#training the model
SLR.fit(x,y)
#y^
prediction=SLR.predict(x)

from sklearn.metrics import mean_squared_error as mse
mse=(y,prediction)
plt.plot(x,prediction,color='g',marker='*')

print("MSE: ",mse)
print("intercept: ",SLR.intercept_)
print("slope: ",SLR.coef_)
```

B]Read an excel file to fetch x and y values and repeat linear regression.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


df = pd.read_csv('//content/sample_data/MLLL.csv')


x = df['A']
y = df['B']

#display(df)
plt.scatter(x, y)
plt.title("Linear Regression")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()

# Prepare
SLR = LinearRegression()
x_reshaped = x.values.reshape(-1, 1)

# Train the model
SLR.fit(x_reshaped, y)

# Make predictions
predictions = SLR.predict(x_reshaped)
mse_value = mean_squared_error(y, predictions)

# Plot
regression_x = np.linspace(min(x), max(x), 100)
regression_y = SLR.predict(regression_x.reshape(-1, 1))

plt.scatter(x, y)
plt.plot(regression_x, regression_y, color='g', marker='*', label='Regression Line')
plt.title("Linear Regression")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()

print("MSE:", mse_value)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```
